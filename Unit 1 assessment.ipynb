```python
def add_expense(expenses_list):
    # Ask the user for all details
    print("Hey this is the expense tracker")
    date = input("Please provide the date of the transaction (YY-MM-DD):")
    while (date == ""):
        print("date cannot be blank!")
        date = input("Try again") 
    print("The date was", date)
    category = input("Please enter the category (e.g., Food, Travel):")
    while (category == ""):
        print("category cannot be blank!")
        category = input("Try again") 
    print("The category was", category)
    
    while True:
        amount_input = input("Please enter the amount: ")
        if amount_input == "":
            print("Amount cannot be blank!")
            continue
        try:
            amount = float(amount_input)
            print("The amount spent was", amount)
            break
        except ValueError:
            print("Amount must be a valid number!")
            
    description = input("Please provide a brief description")
    while (description == ""):
        print("description cannot be blank!")
        description = input("Try again") 
    print("The description was", description)

    # Create a dictionary for this expense
    expense = {
        'date': date,
        'category': category,
        'amount': amount,
        'description': description
    }

    # Add it to the provided list
    expenses_list.append(expense)

    print("Expense added successfully!")
    
    # Create a list to hold all expenses
expenses = []

    # Call the function
add_expense(expenses)
    # Display whatâ€™s stored
print(expenses)
```

```python
def set_monthly_budget():
    while True:
        budget_input = input("Enter the total amount you want to budget for the month: ")
        if budget_input == "":
            print("Budget cannot be blank!")
            continue
        try:
            budget = float(budget_input)
            return budget
        except ValueError:
            print("Budget must be a valid number!")

def calculate_total_expenses(expenses_list):
    total = 0
    for expense in expenses_list:
        total += expense['amount']
    return total

def compare_budget(expenses_list, budget):
    total_spent = calculate_total_expenses(expenses_list)

    print("Total expenses so far:", total_spent)
    print("Monthly budget:", budget)

    if total_spent > budget:
        print("You have exceeded your budget!")
    else:
        remaining = budget - total_spent
        print("You have", remaining, "left for the month")
        
monthly_budget = set_monthly_budget()
compare_budget(expenses, monthly_budget)
```

```python
import csv
import os

def save_expenses_to_csv(expenses_list, filename="expenses.csv"):
    with open(filename, mode="w", newline="") as file:
        writer = csv.writer(file)

        # Header row
        writer.writerow(["date", "category", "amount", "description"])

        # Expense rows
        for expense in expenses_list:
            writer.writerow([
                expense["date"],
                expense["category"],
                expense["amount"],
                expense["description"]
            ])

    print("Expenses saved successfully.")


def load_expenses_from_csv(expenses_list, filename="expenses.csv"):
    if not os.path.exists(filename):
        return  # No file yet, nothing to load

    with open(filename, mode="r", newline="") as file:
        reader = csv.DictReader(file)

        for row in reader:
            expense = {
                "date": row["date"],
                "category": row["category"],
                "amount": float(row["amount"]),
                "description": row["description"]
            }
            expenses_list.append(expense)

    print("Previous expenses loaded.")

save_expenses_to_csv(expenses)
load_expenses_from_csv(expenses)
```

```python
def menu():
    expenses = []

    # Load saved expenses when program starts
    load_expenses_from_csv(expenses)

    while True:
        print("\n--- Expense Tracker Menu ---")
        print("1. Add expense")
        print("2. View expenses")
        print("3. Track budget")
        print("4. Save expenses")
        print("5. Exit")

        choice = input("Choose an option (1-5): ")

        if choice == "1":
            add_expense(expenses)

        elif choice == "2":
            if not expenses:
                print("No expenses recorded yet.")
            else:
                for expense in expenses:
                    print(expense)

        elif choice == "3":
            monthly_budget = set_monthly_budget()
            compare_budget(expenses, monthly_budget)

        elif choice == "4":
            save_expenses_to_csv(expenses)

        elif choice == "5":
            save_expenses_to_csv(expenses)
            print("Goodbye! Your expenses have been saved.")
            break

        else:
            print("Invalid option. Please choose a number between 1 and 5.")
            
menu()
```
